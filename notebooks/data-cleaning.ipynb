**We try to transform the values of each column**

```python
data['Storage'].unique()
```

```python
data['Storage'] = data['Storage'].str.replace(r'[^0-9]', '', regex=True).astype(int)
```

```python
data["RAM"].unique()
```

```python
data['RAM'] = data['RAM'].astype(str).str.extract(r'(\d+)').astype(int)
```

```python
data.head(15)
```

```python
data["Camera"].dtype
```

```python
data["Camera"].unique()
```

```python
def sum_camera_values(x):
    nums = re.findall(r'\d+\.?\d*', str(x))     # extract numbers (int + float)
    return sum(float(n) for n in nums)

data["Camera_TotalMP"] = data["Camera"].apply(sum_camera_values)
```

```python
data
```

```python
data["Camera_Count"] = data["Camera"].str.findall(r'\d+\.?\d*').apply(len)
```

```python
data["Has_3D"] = data["Camera"].str.contains("3D", case=False)
```

```python
data["Has_ToF"] = data["Camera"].str.contains("ToF", case=False)
```

```python
data["Screen Size"].unique()
```

```python
data["Price"].unique()
```

```python
data['Price'] = (
    data['Price']
    .astype(str)                         # ensure string
    .str.replace(r'[^0-9\.]', '', regex=True)  # keep only digits + dot
    .astype(float)
)
```

```python
data["Brand"].unique()
```

```python
data = data.drop_duplicates()
```

```python
data.isna().sum()
```

```python
data.loc[:, 'Price_per_MP'] = data['Price'] / data['Camera_TotalMP']
```

```python
data.loc[:, 'Camera_clean'] = (
    data['Camera']
    .astype(str)
    .str.replace("\n", " ")
    .str.replace("\r", " ")
    .str.strip()
)
```

```python
data.loc[:, 'Main_Camera_MP'] = (
    data['Camera_clean']
    .str.extract(r'(\d+\.?\d*)')
    .astype(float)
)
```

```python
data[['Camera', 'Camera_clean', 'Main_Camera_MP']].head(10)
```

```python
# Normalize ALL whitespace including unicode
data.loc[:, 'Camera_clean'] = (
    data['Camera']
    .astype(str)
    .apply(lambda x: re.sub(r'\s+', ' ', x))  # convert ANY whitespace to single space
    .str.strip()
)
```

```python
# Extract main camera MP correctly
data.loc[:, 'Main_Camera_MP'] = (
    data['Camera_clean']
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
)
```
