```python
data['5g_support'] = data['5g_support'].map({'Yes': True, 'No': False}) #chage yes no to boolean true false
```

```python
months_map = {m: i for i, m in enumerate(calendar.month_name) if m}
data['release_month_num'] = data['release_month'].map(months_map)
```

```python
iphone_16_Pro_max= data[data['model'].str.contains("iPhone 16 Pro Max", case=False, na=False)]
```

```python
avg_price_by_storage = iphone_16_Pro_max.groupby('storage_gb')['price_usd'].mean()
avg_price_by_storage
```

```python
avg_price_by_ram = iphone_16_Pro_max.groupby('ram_gb')['price_usd'].mean()
```

```python
avg_price_by_ram
```

```python
data['clean_model'] = data['model'].str.replace(r"\s\d+$", "", regex=True)
```
