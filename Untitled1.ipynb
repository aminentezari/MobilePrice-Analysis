```python
data['Price_per_RAM'] = data['Price'] / data['RAM']
```

```python
data.groupby("Brand")["Price"].mean().sort_values()
```

**Price increase per 1 GB RAM by brand**

```python
ram_effect = {}

for brand, dfb in data.groupby("Brand"):
    model = LinearRegression()
    X = dfb[['RAM']]
    y = dfb['Price']
    model.fit(X, y)
    ram_effect[brand] = model.coef_[0]   # price increase per 1GB RAM

ram_effect
```

**Price increase per +128GB Storage per brand**

```python
storage_effect = {}

for brand, dfb in data.groupby("Brand"):
    model = LinearRegression()
    X = dfb[['Storage']]
    y = dfb['Price']
    model.fit(X, y)
    storage_effect[brand] = model.coef_[0]   # price increase per 1GB storage

storage_effect
```

```python
ram_effect_df = (
    pd.Series(ram_effect, name='Price_per_GB_RAM')
      .reset_index()
      .rename(columns={'index': 'Brand'})
      .sort_values('Price_per_GB_RAM', ascending=False)
)

ram_effect_df
```

```python
plt.figure(figsize=(8,5))
plt.bar(ram_effect_df['Brand'], ram_effect_df['Price_per_GB_RAM'])
plt.xticks(rotation=45)
plt.ylabel('Price increase per +1 GB RAM')
plt.title('RAM price effect by brand')
plt.tight_layout()
plt.show()
```
