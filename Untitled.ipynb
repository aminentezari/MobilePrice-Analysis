```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
clean_data = pd.read_csv("data/cleaned/mobile_prices_cleaned.csv")
clean_data.head()
```

```python
numeric_cols = [
    'price_usd', 'ram_gb', 'storage_gb', 
    'camera_mp', 'battery_mah', 
    'display_size_inch', 'charging_watt', 'rating'
]

plt.figure(figsize=(10,7))
sns.heatmap(clean_data[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
