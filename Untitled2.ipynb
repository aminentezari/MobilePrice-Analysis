```python
important_cols = [
    'Storage',
    'Price_per_MP',
    'Camera_TotalMP',
    'RAM',
    'Brand',
    'Battery Capacity',
    'Screen Size',
    'Price'
]

df_imp = df[important_cols].copy()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_imp['Brand'] = le.fit_transform(df_imp['Brand'])
```

```python
df_imp = df_imp.apply(pd.to_numeric, errors='coerce')
```

```python
def iqr_mask(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)
```

```python
for col in df_imp.columns:
    df_imp[f'{col}_IQR_outlier'] = iqr_mask(df_imp[col])
```

```python
iqr_outliers = df_imp[df_imp.filter(like='_IQR_outlier').any(axis=1)]
iqr_outliers.head()
```

```python
from sklearn.ensemble import IsolationForest

X = df_imp[important_cols]   # only important columns

iso = IsolationForest(
    contamination=0.05,
    random_state=42
)

df_imp['IF_flag'] = iso.fit_predict(X)
df_imp['IF_outlier'] = df_imp['IF_flag'] == -1
```

```python
isolation_outliers = df_imp[df_imp['IF_outlier']]
isolation_outliers.head()
```

```python
df_imp['combined_outlier'] = df_imp.filter(like='_IQR_outlier').any(axis=1) | df_imp['IF_outlier']

combined_outliers = df_imp[df_imp['combined_outlier']]
combined_outliers.head()
```

```python
df_imp = df[important_cols].copy()
```
